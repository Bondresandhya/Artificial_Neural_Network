# BINARY

## Step Activation Function
#### The Step activation function is used in the perceptron network. This is usually used in single-layer networks to convert to an output that is binary (0 or 1) or Bipolar (-1 or 1). These are called Binary Step Function and Bipolar Step Function Respectively. Here if the input value to a function is greater than a threshold value then the output is 1 (neuron will fire) otherwise 0 or -1 in the case of Bipolar Step Function (neuron will not fire).

```python
def step(x):
    return np.where(x >= 0, 1, 0)

x = np.linspace(-v, v, 100)
y = step(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Step Activation Function")
plt.grid()
plt.ylim(-0.1, 1.1)
plt.show()
```

```python
def linear_activation(x):
    return x

x = np.linspace(-v, v, 100)
y = linear_activation(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Linear Activation Function")
plt.grid()
plt.show()
```

## Non-Linear Activation Functions
#### The non-linear functions are known to be the most used activation functions. It makes it easy for a neural network model to adapt with a variety of data and to differentiate between the outcomes.

#### These functions are mainly divided basis on their range or curves:

## 1) Sigmoid Activation Functions
#### Sigmoid takes a real value as the input and outputs another value between 0 and 1. The sigmoid activation function translates the input ranged in (-∞,∞) to the range in (0,1)

```python
def sigmoid(b,x):
    return 1 / (1 + np.exp(-b*x))
b=float(input("Enter the value of lambda(between 0 to 1): "))
y = np.linspace(-v, v, 100)
x = sigmoid(b,y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Sigmoid Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()


x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=1 / (1 + np.exp(-b*x))
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Sigmoid Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 2)ReLU 

#### This is a more modern and widely used activation function. It stands for Rectified Linear Unit and looks like this:
#### 3Some problems with sigmoid and Hyperbolic Tan (tanh) activation functions such as Vanishing Gradient Problem and Computational Expensive Problem.

#### The RELU activation function returns 0 if the input value to the function is less than 0 but for any positive input, the output is the same as the input. It is also continuous but non-differentiable at 0 and at values less than 0 because its derivative is 0 for any negative input.



```python
def relu(x):
    return np.maximum(0, x)

x = np.linspace(-v, v, 100)
y = relu(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("ReLU Activation Function")
plt.grid()
plt.ylim(-1, 11)
plt.show()


x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.maximum(0, x)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of ReLU Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 3)Tanh 
#### This activation function is somewhat similar to the sigmoid in the sense that it also maps the input values to an s-shaped curve, but in this case, the range is (-1, 1) and is centered at 0, which solves one of the problems with the sigmoid function. Tanh stands for the hyperbolic tangent, which is just the hyperbolic sine divided by the hyperbolic cosine, similarly to the regular tangent.

```python
def tanh(x):
    return np.tanh(x)
    
x = np.linspace(-v, v, 100)
y = tanh(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("tanh Activation Function")
plt.grid()
plt.ylim(-1.1, 1.1)
plt.show()


x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.tanh(x)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of tanh Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 4)Signum 
#### The signum function simply gives the sign for the given values of x. For x value greater than zero, the value of the output is +1, for x value lesser than zero, the value of the output is -1, and for x value equal to zero, the output is equal to zero. The signum function can be defined and understood from the below expression.

```python
def signum(x):
    return np.where(x >= 0, 1, -1)

x = np.linspace(-v, v, 100)
y = signum(x)

plt.plot(x, y)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Signum Activation Function")
plt.grid()
plt.ylim(-1.5, 1.5)
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.where(x >= 0, 1, -1)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Signum Activation Function")
plt.ylim(-0.2, 3.0)
plt.grid()
plt.show()
```

## 5)Softplus 
#### The image of the Softplus function is similar to that of the Relu function, but is smoother. Because the Softplus function requires logarithmic operation, it is not as simple as the Rule function in use. In addition, it is experimentally proven that the Rule function performs better than the Softplus function. Its expression is: softplus(x) = ln(1+e^x)

```python
def softplus(x):
    return np.log(1 + np.exp(x))

y = np.linspace(-v, v, 100)
x = softplus(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Softplus Activation Function")
plt.ylim(-0.2, 3.0)
plt.grid()
plt.show()

def softplus_der(x): 
    return np.divide(1.,1.+np.exp(-x))
y = np.linspace(-v, v, 100)
x = softplus_der(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Softplus Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.log(1 + np.exp(x))
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Softplus Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 6)Swish
#### The Swish function [3] is a new type of activation function proposed by Google in 2017, which has caused a huge sensation in its presence. Swish function expressions are not obtained through theoretical reasoning, but through experimental application of small-scale exhaustive search and large-scale RNN controller application. A large number of experiments prove that its effect is much better than the Relu function. Its expression is:

```python
def swish(b, x):
    swish = x * sigmoid(b,x)
    return swish
b=float(input("Enter the value of lambda(between 0 to 1): "))
y = np.linspace(-v, v, 100)
x = swish(b,y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Swish Activation Function")
plt.ylim(-0.9, 4)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=x * sigmoid(b,x)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Swish Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 7)FTS
#### The full name of FTS function [4] is Flatten-T Swish. Hock Hung Chieng proposed the FTS function in 2018 based on the Swish function. The proposition of FTS function was inspired by Swish. At first, the expression was Relu (x)·Sigmoid (x). In this case, when x <0, the value of the function is still 0, as well as the reverse gradient. The inactivity problem is still not solved. Therefore, we decided to add a constant T to Relu (x)·Sigmoid (x). Generally, the value of T is negative, and the neural network benefits from the negative form

```python
def fts(x):
    a1=swish(b,x)+t
    return np.maximum(a1, t)
t1=int(input("Enter a number:"))
t=-t1
b=1
y = np.linspace(-v, v, 100)
x = fts(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("FTS Activation Function")
plt.ylim(-2, 4)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=fts(x)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of FTS Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 8)Relus-softplus
#### The Relus-softplus function [5] was proposed by Qu in 2017 by analyzing the advantages and disadvantages of various traditional activation functions and combining the Relu function with the Softplus function (which can suppress the excessive sparsity of Relu).

```python
def relus_softplus(x):
    return np.where(x>0,x,np.log(1 + np.exp(x))-np.log(2))

y = np.linspace(-v, v, 100)
x = relus_softplus(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Relus-Softplus Activation Function")
plt.ylim(-1.0, 1.2)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.where(x>0,x,np.log(1 + np.exp(x))-np.log(2))
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of FTS Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 9) Leaky ReLU Activation Function

```python
def leaky_relu(x):
    return np.where(x>0,x,x*0.05)
                      
y = np.linspace(-v, v, 100)
x = leaky_relu(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Leaky Relu Activation Function")
plt.ylim(-1.0, 1.2)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.where(x>0,x,x*0.05)
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Leaky Relu Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 10) ELU Activation Function

```python
def elu(x):
    return np.where(x>0,x,1.*(np.exp(x)-1))
y = np.linspace(-v, v, 100)
x = elu(y)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("ELU Activation Function")
plt.ylim(-1.5, 1.0)
plt.grid()
plt.show()

x = np.linspace(-v, v, 100)
dx=x[1]-x[0]
y=np.where(x>0,x,1.*(np.exp(x)-1))
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of elu Activation Function")
plt.ylim(-0.2, 1.2)
plt.grid()
plt.show()
```

## 11)SeLU Activation Function

```python
def selu(x,_lambda,alpha):
    return np.where(x>0,x*_lambda,alpha*_lambda*(np.exp(x)-1))

_lambda=1.05070098
alpha=1.67326324
y = np.linspace(-v, v, 200)
x = selu(y,_lambda,alpha)

plt.plot(y, x)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("SELU Activation Function")
plt.ylim(-2.0, 1.0)
plt.grid()
plt.show()

x = np.linspace(-20, 20, 100)
dx=x[1]-x[0]
y=np.where(x>0,x*_lambda,alpha*_lambda*(np.exp(x)-1))
dydx=np.gradient(y,dx)
plt.plot(x, dydx)
plt.xlabel("Input")
plt.ylabel("Output")
plt.title("Derivative of Selu Activation Function")
plt.ylim(-0.2, 1.8)
plt.grid()
plt.show()
```
